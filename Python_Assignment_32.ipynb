```python
# Q3

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-10, 10, 1000)

pdf1 = norm.pdf(x, loc=0, scale=1)
pdf2 = norm.pdf(x, loc=0, scale=2)
pdf3 = norm.pdf(x, loc=3, scale=1)

plt.plot(x, pdf1, label="Œº=0, œÉ=1", color="blue")
plt.plot(x, pdf2, label="Œº=0, œÉ=2", color="red", linestyle="dashed")
plt.plot(x, pdf3, label="Œº=3, œÉ=1", color="green", linestyle="dotted")

plt.xlabel("X")
plt.ylabel("Probability Density")
plt.title("Effect of Mean (Œº) and Standard Deviation (œÉ) on Normal Distribution")
plt.legend()
plt.grid()
plt.show()

```

```python
# Q4

""" The normal distribution is one of the most important probability distributions in statistics, data science, and real-world applications. Its importance comes from the Central Limit Theorem (CLT), which states that the sum or average of a large number of independent random variables tends to follow a normal distribution, regardless of the original distribution of the variables.

Why is the Normal Distribution Important?
Common in Nature and Real Life

Many natural phenomena (e.g., human height, IQ, blood pressure) follow a normal distribution, making it a useful model for analyzing real-world data.

Foundation for Inferential Statistics

Many statistical tests, such as t-tests, ANOVA, and regression analysis, assume data follows a normal distribution.

Central Limit Theorem (CLT)

Even if individual data points are not normally distributed, the mean of large samples from any distribution tends to be normally distributed.

Simplifies Statistical Calculations

The properties of the normal curve allow for easy probability calculations using Z-scores and percentiles.

Used in Machine Learning & Data Science

Many algorithms, like Linear Regression and Gaussian Naive Bayes, assume that data follows a normal distribution.

Real-Life Examples of Normal Distribution
Here are a few real-world examples where the normal distribution is observed:

Human Heights

If you measure the heights of a large population, most people will have an average height, with fewer very tall or very short individuals.

Example: The height of adult males in the U.S. follows a normal distribution with a mean of 5'9" (175 cm) and a standard deviation of about 3 inches (7.6 cm).

IQ Scores

IQ scores follow a normal distribution with a mean of 100 and a standard deviation of 15.

About 68% of people have IQs between 85 and 115, and about 95% have IQs between 70 and 130.

Blood Pressure Readings

Normal blood pressure readings in adults follow a bell-shaped curve, with an average systolic pressure around 120 mmHg.

Stock Market Returns

In financial modeling, daily stock returns are often assumed to be normally distributed.

While not always true, this assumption is useful for risk assessment and portfolio optimization.

Exam Scores

In large standardized exams (e.g., SAT, GRE), student scores tend to be normally distributed around the mean score.

Measurement Errors

Errors in scientific measurements (due to random fluctuations) typically follow a normal distribution. """

```

```python
# Q6

import scipy.stats as stats

mu = 50
sigma = 10
x_value = 60

probability = 1 - stats.norm.cdf(x_value, loc=mu, scale=sigma)

print(f"Probability that X > 60: {probability:.4f}")
```

```python
# Q7

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

a, b = 0, 30

x = np.linspace(a, b, 1000)
pdf_values = uniform.pdf(x, loc=a, scale=b-a)

plt.plot(x, pdf_values, label="Uniform Distribution (0 to 30 min)", color="blue")
plt.fill_between(x, pdf_values, alpha=0.3, color="blue")
plt.xlabel("Time (minutes)")
plt.ylabel("Probability Density")
plt.title("Continuous Uniform Distribution Example")
plt.legend()
plt.grid()
plt.show()
```

```python
# Q9

"""The Central Limit Theorem (CLT) is a fundamental statistical concept that describes the behavior of the sampling distribution of the sample mean (or sum) when independent
random variables are added together. It states that, regardless of the original population‚Äôs distribution, the distribution of the sample mean will approximate a normal (Gaussian)
distribution as the sample size becomes large, provided certain conditions are met.

The Central Limit Theorem is significant for several reasons:

1. Foundation for Inferential Statistics
The CLT provides a theoretical basis for making inferences about population parameters using sample statistics. Since many statistical methods assume normality,
 the CLT ensures that we can apply these methods even when dealing with non-normal populations, as long as the sample size is sufficiently large.
2. Sampling Distribution Approximation
It allows us to approximate the sampling distribution of the sample mean with a normal distribution when
n
 is large. This approximation holds true regardless of whether the underlying population is normally distributed or not.
3. Practical Applications in Real-World Problems
In real-world scenarios where data often do not follow a perfect normal distribution (e.g., skewed or multimodal distributions), the CLT ensures that we can still use tools like
 confidence intervals and hypothesis tests effectively by relying on large enough samples.
4. Simplification of Complex Problems
By approximating non-normal distributions with a normal one through sampling, it simplifies complex probability calculations and makes them computationally feasible.
5. Key Role in Quality Control and Risk Management
In fields such as manufacturing or finance, where averages or sums of measurements are analyzed over time, understanding that these averages tend toward normality helps in setting
 thresholds for quality control or assessing risks.
6. Basis for Large Sample Theory
Many advanced statistical techniques rely on asymptotic properties derived from the CLT to justify their validity when applied to large datasets."""
```

```python
# Q10

"""Assumptions of the Central Limit Theorem :
The Central Limit Theorem (CLT) is a fundamental concept in statistics that relies on several key assumptions to ensure its validity. These assumptions are crucial for applying
the theorem correctly and making accurate statistical inferences based on sample data. Here are the main assumptions:

Random Sampling: The samples must be randomly selected from the population. This ensures that every member of the population has an equal chance of being included in the sample,
which helps to create a representative sample.

Independence: The observations within the sample must be independent of each other. This means that the selection or outcome of one observation does not influence another
observation, ensuring that each sample is a true reflection of random variation.

Sample Size: The sample size should be sufficiently large for the CLT to apply. A common rule of thumb is that a sample size of 30 or more is generally considered adequate for
the CLT to hold true, although this can vary depending on the underlying distribution.

Finite Variance: The population from which samples are drawn must have a finite variance. If the population distribution has infinite variance (as seen in certain distributions
like Cauchy), then the CLT may not apply. """
```

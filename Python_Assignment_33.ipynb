```python
# Q1

""" Estimation Statistics
Estimation statistics is a branch of statistical inference that involves estimating population parameters based on sample data. It provides methods to derive estimates that can
inform researchers about the characteristics of a larger group from which a sample has been drawn. The two primary types of estimation in statistics are point estimation and
interval estimation.

Point Estimate
A point estimate is a single value derived from sample data that serves as a best guess or approximation of an unknown population parameter. For instance, if researchers want to
estimate the average height of adult males in a city, they might take a random sample of 100 adult males and calculate their average height. This calculated average (e.g., 175 cm)
is the point estimate for the population mean height.

Characteristics of Point Estimates
Simplicity: Point estimates provide a straightforward summary measure.
Precision: While they offer a specific value, point estimates do not convey information about the uncertainty associated with that estimate.
Bias: A point estimate can be biased if it systematically overestimates or underestimates the true population parameter.
Point estimates are often used in various fields, including economics, psychology, and health sciences, where researchers need quick insights into population characteristics without
extensive data collection.

Interval Estimate
An interval estimate, on the other hand, provides a range of values within which the population parameter is expected to lie with a certain level of confidence. This approach
acknowledges the inherent uncertainty in estimating parameters based on sample data and offers more informative insights than point estimates alone.

Confidence Intervals
The most common form of interval estimation is the confidence interval (CI). A confidence interval consists of two parts: an upper limit and a lower limit, which together define
an interval around the point estimate. For example, if researchers calculate that the average height of adult males is 175 cm with a 95% confidence interval ranging from 172 cm to
 178 cm, this means they are 95% confident that the true average height lies within this range."""
```

```python
# Q2

from scipy import stats
import math

def estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence=0.95):

    standard_error = sample_std_dev / math.sqrt(sample_size)
    t_score = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)
    margin_of_error = t_score * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound
sample_mean = 50
sample_std_dev = 10
sample_size = 30
confidence_level = 0.95

ci = estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level)
print(f"Estimated population mean is between {ci[0]:.2f} and {ci[1]:.2f} with {confidence_level*100}% confidence.")
```

```python
# Q6

"""Null and Alternative Hypothesis
In hypothesis testing, we test assumptions about a population using statistical methods. There are two key hypotheses:

Null Hypothesis (𝐻0H 0​)

It represents the default or no effect assumption.

It assumes no difference, no relationship, or no change in the population.

We either reject or fail to reject H0​
based on statistical evidence.

Alternative Hypothesis (H1 or Ha)

It represents the research claim we want to test.

It suggests that there is a difference, effect, or relationship in the population.

If we reject H0​, we provide evidence in favor of H1.

Example : Checking Whether a Coin is Fair H0: The coin is fair, meaning it lands on heads 50% of the time. (P heads=0.5)
H1: The coin is biased towards heads. (P head>0.5)"""
```

```python
from scipy import stats
import math
import numpy as np

def estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence=0.95):
    standard_error = sample_std_dev / math.sqrt(sample_size)
    t_score = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)
    margin_of_error = t_score * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

def hypothesis_test_population_means(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    conclusion = "Reject H0: There is a significant difference between the population means." if p_value < alpha else "Fail to reject H0: No significant difference found."
    return t_stat, p_value, conclusion

def hypothesis_test_coin_fairness(coin_flips, alpha=0.05):
    observed_heads = sum(coin_flips)
    total_flips = len(coin_flips)
    expected_heads = total_flips * 0.5
    std_dev = math.sqrt(total_flips * 0.5 * 0.5)

    z_stat = (observed_heads - expected_heads) / std_dev
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    conclusion = "Reject H0: The coin is biased." if p_value < alpha else "Fail to reject H0: The coin is fair."
    return z_stat, p_value, conclusion

sample1 = np.random.normal(70, 12, 40)
sample2 = np.random.normal(75, 10, 35)

t_stat, p_value, result = hypothesis_test_population_means(sample1, sample2)
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
print(result)

coin_flips = np.random.choice([0, 1], size=100, p=[0.5, 0.5])
z_stat, p_value, coin_result = hypothesis_test_coin_fairness(coin_flips)
print(f"Z-statistic: {z_stat:.4f}, P-value: {p_value:.4f}")
print(coin_result)
```

```python
# Q11
""" Student’s t distribution, commonly referred to as the t distribution, is a continuous probability distribution that is symmetric and bell-shaped, similar to the normal
distribution. However, it has heavier tails, which allows for a greater probability of extreme values. This characteristic makes the t distribution particularly useful when dealing
with small sample sizes or when the population variance is unknown.

The t distribution was first introduced by William Sealy Gosset in 1908 under the pseudonym “Student.” It was developed to address the challenges associated with small sample sizes
 in statistical analysis, particularly in estimating population parameters such as means. The shape of the t distribution is determined by its degrees of freedom (df), which are
  calculated as the sample size minus one (n - 1). As the degrees of freedom increase, the t distribution approaches the standard normal distribution.

When to Use the t Distribution
The t distribution is primarily used in situations where:

Small Sample Sizes: The sample size is typically less than 30. In these cases, using the normal distribution may not be appropriate due to increased variability.

Unknown Population Variance: The population standard deviation is unknown. Instead of using this unknown value, researchers estimate it from their sample data.

Approximately Normal Data: The data should be approximately normally distributed. If this assumption holds true, then using the t distribution will yield valid results.

Hypothesis Testing: It is commonly used in hypothesis testing scenarios such as:

One-sample t-tests to compare a sample mean against a known population mean.
Two-sample t-tests to compare means between two independent groups.
Paired sample t-tests for comparing means from related groups.
Confidence Intervals: The t distribution is also utilized for constructing confidence intervals around estimated means when dealing with small samples and unknown population
variances."""
```

```python
# Q14

from scipy import stats
import math
import numpy as np
import matplotlib.pyplot as plt
def estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence=0.95):
    standard_error = sample_std_dev / math.sqrt(sample_size)
    t_score = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)
    margin_of_error = t_score * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

def hypothesis_test_population_means(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    conclusion = "Reject H0: There is a significant difference between the population means." if p_value < alpha else "Fail to reject H0: No significant difference found."
    return t_stat, p_value, conclusion

def hypothesis_test_coin_fairness(coin_flips, alpha=0.05):
    observed_heads = sum(coin_flips)
    total_flips = len(coin_flips)
    expected_heads = total_flips * 0.5
    std_dev = math.sqrt(total_flips * 0.5 * 0.5)

    z_stat = (observed_heads - expected_heads) / std_dev
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    conclusion = "Reject H0: The coin is biased." if p_value < alpha else "Fail to reject H0: The coin is fair."
    return z_stat, p_value, conclusion

def plot_students_t_distribution(df=10):
    x = np.linspace(-4, 4, 1000)
    y = stats.t.pdf(x, df)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f"t-distribution (df={df})", color='blue')
    plt.title("Student's t-Distribution")
    plt.xlabel("t-value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

def two_sample_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    conclusion = "Reject H0: There is a significant difference between the population means." if p_value < alpha else "Fail to reject H0: No significant difference found."
    return t_stat, p_value, conclusion

def test_new_drug_effect(sample_mean, expected_mean, sample_std_dev, sample_size, alpha=0.05):
    standard_error = sample_std_dev / math.sqrt(sample_size)
    t_stat = (sample_mean - expected_mean) / standard_error
    p_value = stats.t.sf(abs(t_stat), df=sample_size - 1) * 2

    conclusion = "Reject H0: The drug effect is significantly different from 10 mmHg." if p_value < alpha else "Fail to reject H0: No significant difference found."
    return t_stat, p_value, conclusion

sample1 = np.random.normal(70, 12, 40)
sample2 = np.random.normal(75, 10, 40)

t_stat, p_value, result = two_sample_t_test(sample1, sample2)
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
print(result)

coin_flips = np.random.choice([0, 1], size=100, p=[0.5, 0.5])
z_stat, p_value, coin_result = hypothesis_test_coin_fairness(coin_flips)
print(f"Z-statistic: {z_stat:.4f}, P-value: {p_value:.4f}")
print(coin_result)
plot_students_t_distribution(df=10)
sample_mean_revenue = 500
sample_std_dev_revenue = 50
sample_size_revenue = 50
confidence_interval = estimate_population_mean(sample_mean_revenue, sample_std_dev_revenue, sample_size_revenue, confidence=0.95)
print(f"95% Confidence Interval for the population mean revenue: {confidence_interval}")

sample_mean_bp = 8
expected_mean_bp = 10
sample_std_dev_bp = 3
sample_size_bp = 100

t_stat, p_value, drug_result = test_new_drug_effect(sample_mean_bp, expected_mean_bp, sample_std_dev_bp, sample_size_bp)
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
print(drug_result)
```

```python
# Q16

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence=0.95):

    standard_error = sample_std_dev / np.sqrt(sample_size)
    t_score = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)
    margin_of_error = t_score * standard_error
    return sample_mean - margin_of_error, sample_mean + margin_of_error

def test_study_material_effect(n1, mean1, std1, n2, mean2, std2, alpha=0.01):

    standard_error = np.sqrt((std1**2 / n1) + (std2**2 / n2))
    t_stat = (mean1 - mean2) / standard_error
    df = min(n1 - 1, n2 - 1)
    p_value = stats.t.sf(abs(t_stat), df) * 2
    conclusion = "Reject H0: The population means for the two groups are significantly different." if p_value < alpha else "Fail to reject H0: No significant difference found."
    return t_stat, p_value, conclusion, df

def plot_t_distribution(t_stat, df, alpha):

    x = np.linspace(-4, 4, 1000)
    y = stats.t.pdf(x, df)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f"t-distribution (df={df})", color='blue')

    critical_value = stats.t.ppf(1 - alpha/2, df)
    plt.axvline(x=critical_value, color='red', linestyle='dashed', label="Critical value (+)")
    plt.axvline(x=-critical_value, color='red', linestyle='dashed', label="Critical value (-)")

    plt.axvline(x=t_stat, color='green', linestyle='solid', label="Test statistic")

    plt.title("Hypothesis Test: Two-Sample t-Test")
    plt.xlabel("t-value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

group1_n = 30
group1_mean = 80
group1_std = 10
group2_n = 40
group2_mean = 75
group2_std = 8

t_stat, p_value, study_result, df = test_study_material_effect(group1_n, group1_mean, group1_std, group2_n, group2_mean, group2_std)
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
print(study_result)

plot_t_distribution(t_stat, df, alpha=0.01)
```

```python
# Q17

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def test_product_weight(mu, sigma, sample_size, sample_mean, alpha=0.01):
    standard_error = sigma / np.sqrt(sample_size)
    t_stat = (sample_mean - mu) / standard_error
    df = sample_size - 1
    p_value = stats.t.cdf(t_stat, df)  # One-tailed test
    conclusion = "Reject H0: The true mean weight is significantly less than 5 pounds." if p_value < alpha else "Fail to reject H0: No significant difference found."
    return t_stat, p_value, conclusion, df

def plot_product_weight_test(t_stat, df, alpha):
    x = np.linspace(-4, 4, 1000)
    y = stats.t.pdf(x, df)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f"t-distribution (df={df})", color='blue')
    critical_value = stats.t.ppf(alpha, df)  # One-tailed test
    plt.axvline(x=critical_value, color='red', linestyle='dashed', label="Critical value")
    plt.axvline(x=t_stat, color='green', linestyle='solid', label="Test statistic")

    plt.title("Hypothesis Test: One-Sample t-Test for Product Weight")
    plt.xlabel("t-value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()
target_mean = 5
sample_std = 0.5
sample_size = 25
sample_mean = 4.8

t_stat, p_value, weight_result, df = test_product_weight(target_mean, sample_std, sample_size, sample_mean)
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
print(weight_result)
plot_product_weight_test(t_stat, df, alpha=0.01)

def estimate_confidence_interval(sample_mean, sample_std, sample_size, confidence=0.99):
    standard_error = sample_std / np.sqrt(sample_size)
    t_score = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)
    margin_of_error = t_score * standard_error
    return sample_mean - margin_of_error, sample_mean + margin_of_error

sample_mean_ads = 4
sample_std_ads = 1.5
sample_size_ads = 50
confidence_interval = estimate_confidence_interval(sample_mean_ads, sample_std_ads, sample_size_ads, confidence=0.99)
print(f"99% Confidence Interval for Ads Watched: {confidence_interval}")
```

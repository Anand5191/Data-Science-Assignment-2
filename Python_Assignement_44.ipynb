```python
# Q2

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df.columns
```

```python
df['Price'] = df['Price'].replace('[₹,]', '', regex=True).astype(int)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Price'], bins=50, kde=True, color='skyblue')
plt.title("Distribution of Flight Prices")
plt.xlabel("Flight Price (in ₹)")
plt.ylabel("Number of Flights")
plt.grid(True)
plt.show()
```

```python
# Q3

min_price = df['Price'].min()
max_price = df['Price'].max()
price_range = max_price - min_price

print("Minimum Price:", min_price)
print("Maximum Price:", max_price)
print("Price Range:", price_range)
```

```python
# Q4

plt.figure(figsize=(12, 6))
sns.boxplot(x='Airline', y='Price', data=df)
plt.xticks(rotation=45)
plt.title("Flight Prices by Airline")
plt.xlabel("Airline")
plt.ylabel("Flight Price (in ₹)")
plt.tight_layout()
plt.show()
```

```python
# Q6

df['Journey_month'] = pd.to_datetime(df['Date_of_Journey'], dayfirst=True).dt.month
monthly_avg = df.groupby('Journey_month')['Price'].mean().reset_index()
```

```python
# Q11

""" Missing Values Impact on Analysis:
Missing values in important columns (e.g., Rating, Category, Price) can skew your analysis. For example:
If the Rating is missing for several apps, your analysis of average ratings might be incomplete.
Missing Category values might make it harder to analyze trends in ratings by category.

How missing values impact your analysis:

Imputation might be necessary: Filling missing values with the mean or median for numerical columns, or with the mode for categorical columns.
Removing rows with missing values might be an option if the missing values are minimal and won't significantly impact your conclusions.

If too many values are missing in a particular column (like Rating or Category), it might be worth examining why they are missing and whether
they should be excluded from your analysis entirely. """

missing_values = df1.isnull().sum()

missing_values[missing_values > 0]
```

```python
# Q12

def convert_size(size):
    if size.endswith('M'):
        return float(size[:-1])
    elif size.endswith('k'):
        return float(size[:-1])
    else:
        return None
df1['Size'] = df1['Size'].replace('Varies with device', pd.NA)
df1['Size'] = df1['Size'].dropna().apply(convert_size)

df1['Rating'] = pd.to_numeric(df1['Rating'], errors='coerce')
df1 = df1.dropna(subset=['Rating', 'Size'])
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Size', y='Rating', data=df1, alpha=0.5)
plt.title("App Size vs Rating")
plt.xlabel("App Size (in KB)")
plt.ylabel("Rating")
plt.grid(True)
plt.show()
```

```python
# Q13

df1['Price'] = df1['Price'].replace(['Free'], '0')

if df1['Price'].dtype == 'object' and any(df1['Price'].str.contains('$')):
    df1['Price'] = df1['Price'].str.replace('$', '', regex=False).astype(float)
else:

    df1['Price'] = pd.to_numeric(df1['Price'], errors='coerce')

df1 = df1.dropna(subset=['Type', 'Price'])
avg_price_by_type = df1.groupby('Type')['Price'].mean().reset_index()
```

```python
plt.figure(figsize=(6, 5))
sns.barplot(x='Type', y='Price', data=avg_price_by_type, palette='pastel')
plt.title("Average App Price by Type")
plt.xlabel("App Type")
plt.ylabel("Average Price ($)")
plt.grid(axis='y')
plt.show()
```

```python
# Q17

df1['Last Updated'] = pd.to_datetime(df1['Last Updated'], errors='coerce')

df1['Year'] = df1['Last Updated'].dt.year
df1['Month'] = df1['Last Updated'].dt.month
df1['Month_Name'] = df1['Last Updated'].dt.strftime('%B')
```

```python
monthly_counts = df1['Month_Name'].value_counts().reindex([
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
])

plt.figure(figsize=(10,5))
sns.barplot(x=monthly_counts.index, y=monthly_counts.values, palette='coolwarm')
plt.xticks(rotation=45)
plt.title("Number of App Updates per Month")
plt.xlabel("Month")
plt.ylabel("Number of Apps Updated")
plt.tight_layout()
plt.show()
```

```python
avg_installs = df1.groupby('Month_Name')['Installs'].mean().reindex(monthly_counts.index)

plt.figure(figsize=(10,5))
sns.lineplot(x=avg_installs.index, y=avg_installs.values, marker='o')
plt.title("Average Installs per Month")
plt.xlabel("Month")
plt.ylabel("Average Installs")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
heatmap_data = df1.groupby(['Year', 'Month'])['Installs'].mean().unstack()
plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap='YlGnBu', linewidths=.5)
plt.title("Average Installs by Year and Month")
plt.xlabel("Month")
plt.ylabel("Year")
plt.show()
```
